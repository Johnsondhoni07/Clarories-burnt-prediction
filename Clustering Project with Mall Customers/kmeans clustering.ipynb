```python
#importing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
#importing Dataset
data=pd.read_csv(r"C:\Users\Johnson\Desktop\8.1 Machine Learning A-Z (Codes and Datasets)\Machine Learning A-Z (Codes and Datasets)\Part 4 - Clustering\Section 24 - K-Means Clustering\Python\Mall_Customers.csv")
X=data.iloc[:,[3,4]].values
```

```python
data.head()
```

```python
#checking of wcss to find optimal no of clusters
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=2)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.show()
    
```

```python
#creating model
model=KMeans(n_clusters=5,init='k-means++',random_state=2)
y_model=model.fit_predict(X)
```

```python
print(y_model)
```

```python
#visualising the plot
plt.figure(figsize=(10,10))
plt.scatter(X[y_model==0,0],X[y_model==0,1],s=150,label='cluster 1',cmap='red')
plt.scatter(X[y_model==1,0],X[y_model==1,1],s=150,label='cluster 2',cmap='green')
plt.scatter(X[y_model==2,0],X[y_model==2,1],s=150,label='cluster 3',cmap='blue')
plt.scatter(X[y_model==3,0],X[y_model==3,1],s=150,label='cluster 4',cmap='black')
plt.scatter(X[y_model==4,0],X[y_model==4,1],s=150,label='cluster 4',cmap='gold')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=300,label='centroid',cmap='yellow')
plt.xlabel('Annual Income')
plt.ylabel('Total Amount Spent')
plt.legend()
plt.show()
```
