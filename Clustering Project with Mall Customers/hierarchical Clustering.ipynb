```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster import hierarchy
```

```python
data=pd.read_csv(r"C:\Users\Johnson\Desktop\8.1 Machine Learning A-Z (Codes and Datasets)\Machine Learning A-Z (Codes and Datasets)\Part 4 - Clustering\Section 24 - K-Means Clustering\Python\Mall_Customers.csv")
X=data.iloc[:,[3,4]].values  #used only 2 colums since to visuvalise the plot of clustering
```

```python
data.head()
```

```python
#plotting of dendrogram
dend=hierarchy.dendrogram(hierarchy.linkage(X,method='ward'))
plt.title('DENDROGRAM')
plt.xlabel('customers')
plt.ylabel('Eucledian Distance')
plt.show()
```

```python
#training a hierarichical model on dataset
from sklearn.cluster import  AgglomerativeClustering
hrmodel=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_model=hrmodel.fit_predict(X)
```

```python
#visuvalisation
plt.figure(figsize=(8,8))
plt.scatter(X[y_model==0,0],X[y_model==0,1],s=150,cmap='red')
plt.scatter(X[y_model==1,0],X[y_model==1,1],s=150,cmap='green')
plt.scatter(X[y_model==2,0],X[y_model==2,1],s=150,cmap='black')
plt.scatter(X[y_model==3,0],X[y_model==3,1],s=150,cmap='blue')
plt.scatter(X[y_model==4,0],X[y_model==4,1],s=150,cmap='gold')
plt.xlabel('salary')
plt.ylabel('amount spent')
plt.legend
```
